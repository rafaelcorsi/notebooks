```python
%matplotlib inline
import matplotlib.pyplot as plt

import sympy as sp
import numpy as np

def generateSin(freq, time, fs):
    n = time*fs
    x = np.linspace(0.0, time, n)
    s = np.sin(freq*x*2*np.pi)
    return (x, s)
```

```python
T  = 2     # segundos
fs = 500  # Frequencia de amostragem
time = np.linspace(0,T,T*fs)
t  = sp.Symbol('t')
N  = 10
```

```python
a0     = 1/(2*sp.pi)*sp.integrate(3,(t,0,sp.pi))
an0Num = a0.evalf()
print(a0)
```

```python
an = []
for n in range(1,N):
    an.append(1/sp.pi*sp.integrate(3*sp.cos(n*t), (t,0,sp.pi)))
    
anNum = [x.evalf() for x in an]
print(anNum)
```

```python
bn = []
for n in range(0,N):
    bn.append(1/sp.pi*sp.integrate(3*sp.sin(n*t), (t,0,sp.pi)))
        
bnNum = [x.evalf() for x in bn]
print(bn)
```

```python
ySin = a0.evalf()*np.ones(fs*T)

plt.plot(an0Num*np.ones(len(time)))
for n in range(1,N-1):
    plt.plot(anNum[n]*np.sin(2*np.pi*n*time))
```
